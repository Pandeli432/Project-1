## Adding the Data from the 2020 survey

```python
# Create a new DataFrame to store the processed data with the required columns
final_df = pd.DataFrame(columns=["motivation", "hear_about", "prior_participation", "activities_experience", "class_interest", "which_class"])

# Using panda's .startswith() method to create a list with the relavent column groups
prior_participation_cols = [col for col in df_20.columns if col.startswith("prior_participation")]
extent_cols = [col for col in df_20.columns if col.startswith("extent")]

hear_about_col = "hear_about_ccm"
motivation_col = "motivation_seeking_class"
class_interest_col = "new_class_interest"
which_class_col = "new_class_interest_specific"

# Loop through each row in the original dataframe
for idx, row in df_20.iterrows():
    # Collect prior participation activities into a list
    prior_participation = []
    # Using the list prior_participation_col to loop through each relavent column from the df_20 dataframe
    for col in prior_participation_cols:
        if row[col] == "Yes":
            # Adds the new data into the column and removes the starting string and keeps the ending of it as a titled word as well as replacing
            # the _ with a space.
            prior_participation.append(col.replace("prior_participation_", "").replace("_", " ").title())

    # Collect activities/experience from "extent" columns where there was some impact into a list
    activities_experience = []
    for col in extent_cols:
        if pd.notnull(row[col]) and row[col] != "No Impact":
            activities_experience.append(col.replace("extent_", "").replace("_", " ").title())
    
    # Convert hear_about, motivation, and which_class into lists if they have multiple values
    hear_about = row[hear_about_col].split(";") if pd.notnull(row[hear_about_col]) else []
    motivation = row[motivation_col].split(";") if pd.notnull(row[motivation_col]) else []
    which_class = row[which_class_col].split(";") if pd.notnull(row[which_class_col]) else ["N/A"]
    
    # Create a new row as a DataFrame for concatenation
    new_row = pd.DataFrame({
        "motivation": [motivation if motivation else ["Not specified"]],
        "hear_about": [hear_about if hear_about else ["Not specified"]],
        "prior_participation": [prior_participation if prior_participation else ["None"]],
        "activities_experience": [activities_experience if activities_experience else ["None"]],
        "class_interest": [row[class_interest_col]],
        "which_class": [which_class]
    })
    
    # Concatenate the new row to final_df
    final_df = pd.concat([final_df, new_row], ignore_index=True)

# Save new dataframe to check for errors
final_df.to_csv('final_df.csv', index=False)
```

## Looking at the distribution of which computing class students want to take in the future

### A pie chart will be used as this is likely the best way to visualize the data. Only the final_df will be used as there
### is no necessity to view trends throughtout the years. Matplotlib module was used as an aid/reference

```python
# Creating an empty dictionary that will have the keys as the class name and the value as a tally
classes_of_interest = {}

# creating a for loop that goes through each row
for index, row in final_df.iterrows():
    # specifying which column to look in
    which_class = row["which_class"]
    # creating an if statement to disregard data that is missing
    if which_class != "['N/A']":
        # Creating another for loop since the answers are in a list and tallying the results to the classes_of_interest library.
        for specific_class in which_class:
            if specific_class in classes_of_interest.keys():
                classes_of_interest[specific_class] += 1
            if specific_class not in classes_of_interest.keys():
                classes_of_interest[specific_class] = 1

# Some of the columns had a string of 'N/A' instead of a list. Will be easier to delete it from the dictionary instead of writting more if statements.
del classes_of_interest["N/A"]

# creating a list that will have the names of the class as a label
class_label=[]
for name_of_class in classes_of_interest.keys():
    class_label.append(name_of_class)

# creating another list with more colors added)
color_wheel = my_colors + ["#36F1CD", "#B0A084","#9F9FED"]

fig, wheel_of_classes = plt.subplots()
wheel_of_classes.pie(classes_of_interest.values(), labels=class_label, colors=color_wheel, autopct='%1.1f%%')
wheel_of_classes.set_title("Distribution of Class Interest")

plt.show()
```
